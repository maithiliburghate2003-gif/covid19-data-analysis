```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
a=pd.read_csv("country_wise_latest.csv")
# print(a)
display(a)
```

```python
# How many countries are present in the dataset?
countries = a['Country/Region'].nunique()
print(countries)

# What are the total confirmed COVID-19 cases worldwide?
total_cases = a['Confirmed'].sum()
print(total_cases)

# What is the total number of deaths and recoveries globally?
total_deaths = a['Deaths'].sum()
total_recoveries = a['Recovered'].sum()
print(total_deaths)
print(total_recoveries)


# Which columns represent daily changes in cases?
a['Daily_Cases'] = a['Confirmed'].diff()
a['Daily_Deaths'] = a['Deaths'].diff()
a['Daily_Recoveries'] = a['Recovered'].diff()
print(a)


# List all unique WHO regions in the dataset.
region=a['WHO Region'].unique()
print(region)













```

		Country-Level Analysis

```python
# 1. Country with highest confirmed cases
highest_confirmed = a.loc[a['Confirmed'].idxmax(), 'Country/Region']
print("Highest confirmed cases:", highest_confirmed)
```

```python
# 2. Country with maximum deaths
highest_deaths = a.loc[a['Deaths'].idxmax(), 'Country/Region']
print("Maximum deaths:", highest_deaths)
```

```python
# 3. Country with highest recovered cases
highest_recovered = a.loc[a['Recovered'].idxmax(), 'Country/Region']
print("Highest recovered cases:", highest_recovered)
```

```python
# 4. Country with highest active cases
highest_active = a.loc[a['Active'].idxmax(), 'Country/Region']
print("Highest active cases:", highest_active)
```

```python
# 5. Top 10 countries with highest confirmed cases
a_confirmed = a[['Country/Region', 'Confirmed']].sort_values(by='Confirmed', ascending=False).head(10)
print("\nTop 10 countries by confirmed cases:\n", a_confirmed)
```

```python
# 6. Top 10 countries with highest death counts
a_deaths = a[['Country/Region', 'Deaths']].sort_values(by='Deaths', ascending=False).head(10)
print("\nTop 10 countries by deaths:\n", a_deaths)
```

```python
# 7. Countries with zero new cases
zero_new_cases = a[a['New cases'] == 0]['Country/Region'].tolist()
print("\nCountries with zero new cases:\n", zero_new_cases)
```

```python
# 8. Countries with zero deaths so far
zero_deaths = a[a['Deaths'] == 0]['Country/Region'].tolist()
print("\nCountries with zero deaths:\n", zero_deaths)
```

#Recovery & Fatality Rate Analysis

```python
# Which country has the highest death rate per 100 cases?
highest_death_rate = a.loc[a['Deaths / 100 Cases'].idxmax(), 'Country/Region']
print("Country has the highest death rate per 100 cases:",highest_death_rate)
```

```python
# Which country has the highest recovery rate per 100 cases?
highest_recovery_rate = a.loc[a['Recovered / 100 Cases'].idxmax(), 'Country/Region']
print("Country has the highest recovery rate per 100 cases:",highest_recovery_rate)
```

```python
# Which countries have a death rate greater than 5%?
rate_greater_than_5 = a.loc[a['Deaths / 100 Cases'] >5/100]['Country/Region'].tolist()
print("Countries with a death rate greater than 5%:\n",rate_greater_than_5)
```

```python
# Which countries have a recovery rate above 80%?
rate_above_80 = a.loc[a['Recovered / 100 Cases'] >80/100]['Country/Region'].tolist()
print("Countries have a recovery rate above 80%:",rate_above_80)
```

```python
# Identify countries where Deaths / 100 Recovered is unusually high
countries = a[['Country/Region','Deaths / 100 Recovered']].head(10)
print("Highest confirmed cases:", countries)
```

Weekly Growth Analysis

```python

# Which country has the highest increase in cases in the last week?
country=a.loc[a['1 week change'].idxmax()]['Country/Region']
print("Highest increase in cases in the last week:",country)
```

```python

# Which country has the highest 1-week percentage increase?
country=a.loc[a['1 week % increase'].idxmax()]['Country/Region']
print("Highest 1-week percentage increase:",country)
```

```python
# Which countries experienced a negative or zero weekly change?
countries=a.loc[a['1 week change']<=0]['Country/Region'].tolist()
print("Countries experienced a negative or zero weekly change:",countries)
```

```python
# Compare current confirmed cases vs last week confirmed cases.
a['Confirmed_vs_LastWeek'] = a['Confirmed'] - a['1 week change']
print("Current confirmed cases vs last week confirmed cases:\n",a['Confirmed_vs_LastWeek'])
```

```python
# Which countryâ€™s COVID situation worsened the most in the last week?
country=a.loc[a['1 week % increase'].idxmax()]['Country/Region']
print("Countryâ€™s COVID situation worsened the most in the last week:",country)
```

```python
# How many countries are present in the dataset?

import matplotlib.pyplot as plt

# print(countries)

# Plot
# plt.figure()
d = a['Country/Region'].nunique()
# plt.bar(a["Country/Region"], a["Confirmed"], label="Current Cases")
plt.bar(a["Country/Region"], a["Confirmed"],data=d,
        label="Last Week Cases", alpha=0.7)

plt.xlabel("Country")
plt.ylabel("Confirmed Cases")
plt.title("Current vs Last Week Confirmed COVID-19 Cases")
plt.xticks(rotation=90)
plt.legend()
plt.show()
```

```python
# Compare current confirmed cases vs last week confirmed cases.
d=a.groupby(['Confirmed last week'],as_index=False)['Confirmed'].sum().sort_values(by='Confirmed',ascending=False).head(10)
ax=sns.barplot(x='Confirmed',y='Confirmed last week',data=d,hue='Confirmed')
for i in ax.containers:
    ax.bar_label(i)
plt.show()
```
